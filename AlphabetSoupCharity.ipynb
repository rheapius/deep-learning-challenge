## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(['EIN', 'NAME'],axis=1, inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.

print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning

print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
value_counts = application_df['APPLICATION_TYPE'].value_counts()
cutoff_value= 500

application_types_to_replace= value_counts[value_counts < cutoff_value].index.tolist()

# Replace in dataframe

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

print(application_df['CLASSIFICATION'].value_counts())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
value_counts = application_df['CLASSIFICATION'].value_counts()
print(value_counts[value_counts > 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
value_counts = application_df['CLASSIFICATION'].value_counts()
cutoff_value= 1000

classifications_to_replace= value_counts[value_counts < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
#converting 'INCOME_AMT' column to integer

application_df['INCOME_AMT']= pd.to_numeric(application_df['INCOME_AMT'], errors='coerce')
application_df['INCOME_AMT'].fillna(0,inplace=True)
application_df['INCOME_AMT']= application_df['INCOME_AMT'].astype(int)
application_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
encoded_df=pd.get_dummies(application_df, columns=['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION',	'USE_CASE','ORGANIZATION','STATUS','SPECIAL_CONSIDERATIONS','IS_SUCCESSFUL'])
encoded_df = encoded_df.astype(int)

df_encoded= pd.concat([application_df, encoded_df], axis=1)
df_encoded.drop(['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION',	'USE_CASE','ORGANIZATION','STATUS','SPECIAL_CONSIDERATIONS','IS_SUCCESSFUL','IS_SUCCESSFUL_0'], axis=1, inplace= True)

df_encoded = df_encoded.loc[:, ~df_encoded.columns.duplicated()]

df_encoded.info()
```

```python
# Split our preprocessed data into our features and target arrays
X= df_encoded.drop('IS_SUCCESSFUL_1', axis=1)
y= df_encoded['IS_SUCCESSFUL_1']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=6, activation="relu", input_dim=36))

#Second hidden layer
nn.add(tf.keras.layers.Dense(units=6, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
print("Model exported to model.h5")
```
